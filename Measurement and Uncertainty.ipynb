```python
from random import random
import numpy
from numpy import mean, arange, std, exp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def computeEntity(p):
    if random() < p:
        return 1
    else:
        return 0
    
def bernouliMeasurement(iterations, sampleSize, probability):
    x=[]
    for i in range(iterations):
        y=[]
        for j in range(sampleSize):
            y.append(computeEntity(probability))
        x.append(round(mean(y),3))
    return x

def bernoulieProb(sampleSize):
    resultX=[]
    resultY=[]
    for p in arange(0.01,0.99, 0.01):
        resultY.append(std(bernouliMeasurement(100, sampleSize, p)))
        resultX.append(p)
    return resultX, resultY

def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))
```

```python
res10=bernouliMeasurement(1000, 10, 0.5);
res100=bernouliMeasurement(1000, 100, 0.5);
```

```python
plt.hist(res10)
plt.hist(res100)
plt.show()
```

```python
x, y = bernoulieProb(100)
```

```python
plt.plot(x, y)
plt.ylabel('Standard Deviation')
plt.xlabel('Probability')
plt.show()
```
